## 📌 What is it?
Dictionaries are ordered (Python 3.7+), mutable key-value stores with O(1) average lookup.

## 🧠 Why AI Engineers need this
API request/response bodies, model configs, token-to-id mappings, metadata stores — everything in AI engineering uses dicts.

````python
# ── CREATION ──
config = {
    "model": "gpt-4o",
    "temperature": 0.7,
    "max_tokens": 2048,
    "stream": True
}

# dict() constructor
config2 = dict(model="claude-3-5-sonnet", temperature=0.5)
print(config["model"])     # gpt-4o
print(config2["model"])    # claude-3-5-sonnet
````

````python
# ── CRUD OPERATIONS ──
d = {"a": 1, "b": 2}

# Read
print(d["a"])          # 1
print(d.get("c", 0))  # 0 — safe get with default (never KeyError)

# Update
d["b"] = 99
d["c"] = 3              # add new key
d.update({"d": 4, "e": 5})

# Delete
del d["a"]
removed = d.pop("b")    # remove & return
print(d)
print(removed)          # 99
````

````python
# ── ITERATION ──
model_scores = {"GPT-4o": 0.91, "Claude": 0.89, "Gemini": 0.87}

for key in model_scores:
    print(key)

for key, value in model_scores.items():
    print(f"{key}: {value:.2f}")

keys   = list(model_scores.keys())
values = list(model_scores.values())
print(f"Best model: {max(model_scores, key=model_scores.get)}")
````

````python
# ── NESTED DICTS (API response style) ──
api_response = {
    "id": "msg_01",
    "model": "claude-3-5-sonnet",
    "usage": {
        "input_tokens": 512,
        "output_tokens": 128
    },
    "content": [{"type": "text", "text": "Hello!"}]
}

# Safe deep access
input_tokens = api_response.get("usage", {}).get("input_tokens", 0)
output_text  = api_response["content"][0]["text"]
print(f"Input tokens: {input_tokens}")
print(f"Response: {output_text}")
````

````python
# ── DICT TRICKS ──
# Merge two dicts (Python 3.9+)
defaults = {"temperature": 0.7, "max_tokens": 1000}
overrides = {"temperature": 0.0, "model": "gpt-4o"}
merged = defaults | overrides   # overrides wins
print(merged)

# Count occurrences
tokens = ["the", "cat", "the", "dog", "the"]
counts = {}
for t in tokens:
    counts[t] = counts.get(t, 0) + 1
print(counts)   # {'the': 3, 'cat': 1, 'dog': 1}

# Better with defaultdict
from collections import defaultdict, Counter
counts2 = Counter(tokens)
print(counts2.most_common(2))   # [('the', 3), ('cat', 1)]
````

## ⚠️ Common Mistakes
```python
# ❌ KeyError — never access unknown keys directly
d = {}
val = d["missing"]   # KeyError!

# ✅ Use .get()
val = d.get("missing", "default")

# ❌ Iterating and modifying simultaneously
for k in d:
    del d[k]   # RuntimeError!

# ✅ Iterate a copy
for k in list(d.keys()):
    del d[k]
```
