## 📌 What is it?
Variables are named containers that store data. Python is *dynamically typed* — you don't declare types; Python infers them at runtime.

## 🧠 Why AI Engineers need this
Every ML model input/output is a Python object. Knowing types prevents silent bugs when feeding data into NumPy arrays, model APIs, or JSON payloads.

## ✏️ Syntax
```python
variable_name = value
type(variable_name)   # check type
type_name(value)      # cast to type
```

## 💡 Examples

````python
# ── INTEGERS ──
age = 30
batch_size = 64
max_tokens = 4096

print(type(age))          # <class 'int'>
print(type(batch_size))   # <class 'int'>
````

````python
# ── FLOATS ──
learning_rate = 0.001
temperature = 0.7        # LLM sampling temperature
dropout = 0.5

print(type(learning_rate))   # <class 'float'>
print(f"LR: {learning_rate:.4f}")   # LR: 0.0010
````

````python
# ── STRINGS ──
model_name = "gpt-4o"
api_key = "sk-..."
prompt = "Summarize the following text:"

print(type(model_name))   # <class 'str'>
print(len(prompt))        # 32
````

````python
# ── BOOLEANS ──
is_training = True
use_gpu = False
stream = True            # for LLM streaming responses

print(type(is_training))  # <class 'bool'>
print(int(is_training))   # 1  — booleans are ints in Python!
````

````python
# ── NONE TYPE ──
response = None           # placeholder before API call
embedding = None

print(type(response))     # <class 'NoneType'>
print(response is None)   # True  (always use 'is', not ==)
````

````python
# ── TYPE CASTING ──
tokens = "1500"           # came in as string from JSON
token_int = int(tokens)   # → 1500
token_float = float(tokens) # → 1500.0

score = 0.987
score_pct = f"{score * 100:.1f}%"   # → "98.7%"
score_str = str(score)               # → "0.987"

print(token_int + 500)   # 2000
print(score_pct)          # 98.7%
````

````python
# ── MULTIPLE ASSIGNMENT ──
x = y = z = 0
a, b, c = 1, 2, 3
first, *rest = [10, 20, 30, 40]   # starred unpacking

print(a, b, c)   # 1 2 3
print(first)     # 10
print(rest)      # [20, 30, 40]
````

## ⚠️ Common Mistakes

```python
# ❌ Using == to check None
if response == None: ...  # works but wrong style

# ✅ Use 'is'
if response is None: ...

# ❌ Integer division surprise
result = 5 / 2    # → 2.5 (float division)
result = 5 // 2   # → 2   (floor division)

# ❌ String + int
print("Tokens: " + 1500)    # TypeError!
print(f"Tokens: {1500}")    # ✅ Use f-string
```
