## 📌 What is it?
Strings are immutable sequences of Unicode characters. Python has rich built-in string manipulation.

## 🧠 Why AI Engineers need this
Prompts, system messages, model outputs, document parsing, and tokenization all work with strings. Mastering strings is essential for LLM engineering.

````python
# ── CREATION ──
s1 = 'single quotes'
s2 = "double quotes"
s3 = """Triple quotes
for multi-line strings — great for system prompts!"""
s4 = r"C:\Users\name"   # raw string (no escape processing)

print(s3)
print(repr(s4))   # shows the actual backslashes
````

````python
# ── INDEXING & SLICING ──
text = "Hello, AI Engineer!"
#       0123456789...

print(text[0])      # H
print(text[-1])     # !
print(text[7:9])    # AI
print(text[:5])     # Hello
print(text[-9:])    # Engineer!
print(text[::2])    # Hlo,A nie r

# Reverse a string
print(text[::-1])   # !reenigné IA ,olleH
````

````python
# ── F-STRINGS (use these!) ──
model = "claude-3-5-sonnet"
tokens = 1423
cost = 0.00285

print(f"Model: {model}")
print(f"Tokens used: {tokens:,}")          # 1,423
print(f"Cost: ${cost:.4f}")                # $0.0029
print(f"Cost per token: ${cost/tokens:.6f}")

# Multi-line f-string (triple-quote style)
system_prompt = (
    f"You are a helpful assistant using {model}.\n"
    f"Max tokens: {tokens}\n"
    "Date: 2025-01-01"
)
print(system_prompt)
````

````python
# ── KEY STRING METHODS ──
text = "  Hello, World!  "

print(text.strip())             # "Hello, World!" — remove whitespace
print(text.upper())             # "  HELLO, WORLD!  "
print(text.lower())             # "  hello, world!  "
print("hello".capitalize())    # "Hello"
print("Hello World".split())   # ['Hello', 'World']
print(",".join(["a","b","c"])) # "a,b,c"
print("hello".replace("l","r"))# "herro"
print("hello world".find("world"))  # 6
print("hello".startswith("he"))     # True
print("hello".endswith("lo"))       # True
print("  ".isspace())               # True
````

````python
# ── AI ENGINEERING: String manipulation for prompts ──
def build_prompt(user_query: str, context_docs: list[str]) -> str:
    """Build a RAG prompt from query and retrieved docs."""
    context = "\n\n".join([f"[Doc {i+1}]: {doc}" 
                             for i, doc in enumerate(context_docs)])
    return f"""Use the following context to answer the question.

Context:
{context}

Question: {user_query}

Answer:"""

docs = ["Python is a high-level language.", "It was created by Guido van Rossum."]
prompt = build_prompt("What is Python?", docs)
print(prompt)
````

## ⚠️ Common Mistakes
```python
# ❌ String concatenation in loops (slow!)
result = ""
for s in big_list:
    result += s   # creates new string each time O(n²)

# ✅ Use join
result = "".join(big_list)   # O(n)

# ❌ Forgetting strings are immutable
s = "hello"
s[0] = "H"   # TypeError!
```
