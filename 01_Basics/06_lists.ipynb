## 📌 What is it?
Lists are ordered, mutable, heterogeneous sequences. They are one of Python's most-used data structures.

## 🧠 Why AI Engineers need this
Batches, token sequences, training data, model predictions, document collections — almost everything in ML lives in lists.

````python
# ── CREATION ──
empty = []
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True, None]   # heterogeneous!
nested = [[1,2,3], [4,5,6], [7,8,9]]    # 2D data

# From other iterables
from_range = list(range(10))
from_string = list("hello")   # ['h','e','l','l','o']
print(from_range)
print(from_string)
````

````python
# ── INDEXING & SLICING ──
data = [10, 20, 30, 40, 50, 60]

print(data[0])    # 10    — first
print(data[-1])   # 60    — last
print(data[1:4])  # [20, 30, 40]
print(data[::2])  # [10, 30, 50]  every other
print(data[::-1]) # [60, 50, 40, 30, 20, 10] reversed

# Batch slicing
batch_size = 2
for i in range(0, len(data), batch_size):
    batch = data[i:i+batch_size]
    print(f"Batch: {batch}")
````

````python
# ── CORE METHODS ──
items = ["a", "b", "c"]

items.append("d")         # add to end     → ['a','b','c','d']
items.insert(1, "X")      # insert at idx  → ['a','X','b','c','d']
items.remove("X")         # remove by val  → ['a','b','c','d']
popped = items.pop()      # remove & return last → 'd'
items.extend(["e", "f"]) # add multiple   → ['a','b','c','e','f']

print(items)
print(len(items))         # 5
print(items.count("a"))   # 1
print(items.index("c"))   # 2
````

````python
# ── SORTING ──
scores = [0.87, 0.95, 0.62, 0.91, 0.78]

# in-place sort
scores.sort(reverse=True)
print(scores)   # [0.95, 0.91, 0.87, 0.78, 0.62]

# new sorted list
original = [3, 1, 4, 1, 5]
ranked = sorted(original, reverse=True)
print(original)  # unchanged
print(ranked)

# sort list of dicts by key
results = [{"model":"A","score":0.91}, {"model":"B","score":0.87}]
results.sort(key=lambda x: x["score"], reverse=True)
print(results[0]["model"])   # A
````

````python
# ── LIST AS A QUEUE / STACK ──
from collections import deque

# Stack (LIFO) — use list
stack = []
stack.append("item1")
stack.append("item2")
print(stack.pop())   # item2

# Queue (FIFO) — use deque for O(1) pops from front
queue = deque()
queue.append("request1")
queue.append("request2")
print(queue.popleft())   # request1
````

## ⚠️ Common Mistakes
```python
# ❌ Shallow copy
a = [1, 2, 3]
b = a       # b is the SAME list!
b.append(4)
print(a)    # [1, 2, 3, 4] — a changed!

# ✅ Copy properly
b = a.copy()       # shallow copy
b = a[:]           # slice copy
import copy
b = copy.deepcopy(a)  # for nested lists
```
