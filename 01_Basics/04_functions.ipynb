## 📌 What is it?
Functions are reusable blocks of code. In Python they're first-class objects — you can pass them as arguments, return them, and store them in variables.

## 🧠 Why AI Engineers need this
Data preprocessing pipelines, custom loss functions, API wrappers, and callback hooks all use functions extensively.

````python
# ── BASIC FUNCTION ──
def greet(name):
    """Return a greeting string."""
    return f"Hello, {name}!"

print(greet("Claude"))   # Hello, Claude!
````

````python
# ── DEFAULT ARGUMENTS ──
def call_llm(prompt, model="gpt-4o", temperature=0.7, max_tokens=1000):
    """Simulate an LLM call with sensible defaults."""
    return {
        "prompt": prompt,
        "model": model,
        "temperature": temperature,
        "max_tokens": max_tokens
    }

# Use defaults
r1 = call_llm("What is RAG?")
print(r1["model"])        # gpt-4o

# Override specific defaults
r2 = call_llm("Write code", temperature=0.0, max_tokens=2000)
print(r2["temperature"])  # 0.0
````

````python
# ── LAMBDA (anonymous function) ──
square = lambda x: x ** 2
print(square(5))   # 25

# Useful for sorting, map, filter
scores = [("modelA", 0.91), ("modelB", 0.87), ("modelC", 0.95)]
scores.sort(key=lambda x: x[1], reverse=True)   # sort by score
print(scores)   # [('modelC', 0.95), ('modelA', 0.91), ('modelB', 0.87)]

# filter out low scores
good = list(filter(lambda x: x[1] > 0.88, scores))
print(good)
````

````python
# ── DOCSTRINGS + TYPE HINTS (best practice) ──
def chunk_text(text: str, chunk_size: int = 500, overlap: int = 50) -> list[str]:
    """
    Split text into overlapping chunks for RAG pipelines.
    
    Args:
        text: Input text to chunk.
        chunk_size: Max characters per chunk.
        overlap: Overlap between chunks to preserve context.
    
    Returns:
        List of text chunks.
    """
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start += chunk_size - overlap
    return chunks

sample = "A" * 1200
chunks = chunk_text(sample, chunk_size=500, overlap=100)
print(f"Number of chunks: {len(chunks)}")
print(f"Chunk 1 length: {len(chunks[0])}")
````

## ⚠️ Common Mistakes
```python
# ❌ Mutable default argument (classic Python bug!)
def add_to_list(item, lst=[]):
    lst.append(item)
    return lst

add_to_list(1)   # [1]
add_to_list(2)   # [1, 2] — NOT [2]! lst persists

# ✅ Use None instead
def add_to_list(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst
```
