## 📌 What is it?
Operators perform operations on values. Python has arithmetic, comparison, logical, assignment, identity, membership, and bitwise operators.

## 🧠 Why AI Engineers need this
Comparisons drive model evaluation logic. Logical operators build conditional data pipelines. Assignment operators update running metrics.

## ✏️ Syntax
```python
+  -  *  /  //  %  **   # arithmetic
==  !=  >  <  >=  <=    # comparison
and  or  not             # logical
is  is not               # identity
in  not in               # membership
+=  -=  *=  /=           # assignment
```

````python
# ── ARITHMETIC ──
a, b = 10, 3
print(a + b)    # 13   addition
print(a - b)    # 7    subtraction
print(a * b)    # 30   multiplication
print(a / b)    # 3.33 true division
print(a // b)   # 3    floor division
print(a % b)    # 1    modulus (remainder)
print(a ** b)   # 1000 exponentiation

# AI engineering use-case
total_cost = 0.002 * (1500 / 1000)   # $0.002 per 1K tokens
print(f"Cost: ${total_cost:.5f}")
````

````python
# ── COMPARISON ──
accuracy = 0.95
threshold = 0.90

print(accuracy > threshold)   # True
print(accuracy == 1.0)        # False
print(accuracy != threshold)  # True

# Chaining (very Pythonic!)
val = 0.7
print(0.0 < val < 1.0)    # True — temperature in valid range
````

````python
# ── LOGICAL ──
has_api_key = True
is_online = True
use_cache = False

if has_api_key and is_online:
    print("Ready to call LLM")     # prints

if use_cache or is_online:
    print("Can proceed")           # prints

if not use_cache:
    print("Will make fresh call")  # prints
````

````python
# ── MEMBERSHIP (very useful in AI!) ──
supported_models = ["gpt-4o", "claude-3-5-sonnet", "gemini-pro"]
requested = "gpt-4o"

if requested in supported_models:
    print(f"{requested} is supported ✅")

stop_words = {"the", "a", "an", "is"}
word = "the"
if word in stop_words:
    print(f"'{word}' is a stop word — skip it")
````

````python
# ── ASSIGNMENT OPERATORS ──
tokens_used = 0
tokens_used += 500    # same as: tokens_used = tokens_used + 500
tokens_used += 300
print(f"Total tokens: {tokens_used}")   # 800

loss = 1.0
loss *= 0.9   # decay loss
print(f"Loss: {loss:.3f}")   # 0.900
````

## ⚠️ Common Mistakes
```python
# ❌ Using = instead of == in condition
if x = 5: ...   # SyntaxError
if x == 5: ...  # ✅

# ❌ Float equality
if 0.1 + 0.2 == 0.3: ...   # False! (floating point)
if abs((0.1+0.2) - 0.3) < 1e-9: ...  # ✅
```
