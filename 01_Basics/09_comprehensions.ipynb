## 📌 What is it?
Comprehensions are concise, readable one-liners to create lists, dicts, and sets from iterables.

## 🧠 Why AI Engineers need this
Data transformation, filtering datasets, building token mappings, and vectorizing operations all benefit from comprehensions — they're faster than explicit loops and more Pythonic.

````python
# ── LIST COMPREHENSIONS ──
# [expression for item in iterable if condition]

# Basic
squares = [x**2 for x in range(10)]
print(squares)   # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# With filter
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)   # [0, 4, 16, 36, 64]

# String processing
words = ["  Hello  ", "  World  ", "  AI  "]
cleaned = [w.strip().lower() for w in words]
print(cleaned)   # ['hello', 'world', 'ai']
````

````python
# ── DICT COMPREHENSIONS ──
# {key: value for item in iterable}

# Build word-to-index mapping (like a tokenizer vocab)
vocab = ["<pad>", "<unk>", "the", "cat", "sat"]
word2idx = {word: idx for idx, word in enumerate(vocab)}
idx2word = {idx: word for word, idx in word2idx.items()}

print(word2idx)   # {'<pad>': 0, '<unk>': 1, ...}
print(idx2word[2])  # 'the'

# Filter low-scoring models
scores = {"GPT-4o": 0.91, "GPT-3.5": 0.72, "Claude": 0.89}
top_models = {m: s for m, s in scores.items() if s >= 0.88}
print(top_models)
````

````python
# ── SET COMPREHENSIONS ──
text = "the quick brown fox jumps over the lazy dog"
unique_chars = {c for c in text if c != ' '}
print(f"Unique chars: {len(unique_chars)}")

unique_words = {w.lower() for w in text.split()}
print(f"Unique words: {len(unique_words)}")
````

````python
# ── NESTED COMPREHENSIONS ──
# Flatten a 2D list
matrix = [[1,2,3], [4,5,6], [7,8,9]]
flat = [x for row in matrix for x in row]
print(flat)   # [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Transpose a matrix
transposed = [[matrix[i][j] for i in range(3)] for j in range(3)]
print(transposed)   # [[1,4,7], [2,5,8], [3,6,9]]
````

````python
# ── GENERATOR EXPRESSIONS (memory efficient!) ──
# Same syntax as list comp but with () — lazy evaluation
import sys

big_list = [x**2 for x in range(10_000)]       # loads into memory
generator = (x**2 for x in range(10_000))       # lazy — no memory

print(f"List size:  {sys.getsizeof(big_list):,} bytes")
print(f"Gen size:   {sys.getsizeof(generator):,} bytes")

# Use in sum — never builds the full list
total = sum(x**2 for x in range(1_000_000))     # memory efficient
print(f"Sum: {total:,}")
````

## ⚠️ Common Mistakes
```python
# ❌ Too complex — unreadable
result = [f(x) for x in [g(y) for y in data if cond1(y)] if cond2(x)]

# ✅ Break it up
filtered = [g(y) for y in data if cond1(y)]
result = [f(x) for x in filtered if cond2(x)]

# ❌ Using list comp when generator is better
total = sum([x**2 for x in range(10**6)])  # builds full list!
total = sum(x**2 for x in range(10**6))    # ✅ generator
```
