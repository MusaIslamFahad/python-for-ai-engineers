## 📌 What is it?
**Tuples** are immutable sequences. **Sets** are unordered collections of unique elements.

## 🧠 Why AI Engineers need this
Tuples represent tensor shapes, (key, value) pairs, and config records. Sets are perfect for deduplication of training data and vocabulary building.

````python
# ── TUPLES ──
model_config = ("gpt-4o", 0.7, 4096)   # (model, temp, tokens)
shape = (batch_size, seq_len, hidden) = (32, 512, 768)

# Unpack
model, temp, tokens = model_config
print(f"{model} | temp={temp} | max_tokens={tokens}")

# Single-element tuple needs trailing comma!
single = (42,)
not_tuple = (42)   # just an int!
print(type(single))    # <class 'tuple'>
print(type(not_tuple)) # <class 'int'>

# Tuples are hashable → can be dict keys or set members
visited = {(1,1), (1,2), (2,1)}
print((1,1) in visited)   # True
````

````python
# ── SETS ──
# Unordered, unique elements only
vocab = {"the", "cat", "sat", "on", "the", "mat"}
print(vocab)   # {'the', 'cat', 'sat', 'on', 'mat'} — 'the' deduped

# Deduplicate a corpus
documents = ["hello world", "hello AI", "hello world"]
unique_docs = list(set(documents))
print(f"Unique docs: {len(unique_docs)}")   # 2
````

````python
# ── SET OPERATIONS ──
train_ids = {1, 2, 3, 4, 5}
test_ids  = {4, 5, 6, 7, 8}

print(train_ids | test_ids)    # union         {1,2,3,4,5,6,7,8}
print(train_ids & test_ids)    # intersection  {4, 5}
print(train_ids - test_ids)    # difference    {1, 2, 3}
print(train_ids ^ test_ids)    # symmetric diff {1,2,3,6,7,8}

# Check for overlap — useful for train/test leakage detection
leaked = train_ids & test_ids
if leaked:
    print(f"⚠️ Data leakage! IDs in both splits: {leaked}")
````

## ⚠️ Common Mistakes
```python
# ❌ Empty set gotcha
empty = {}       # This is an empty DICT, not a set!
empty = set()    # ✅ This is an empty set

# ❌ Sets are unordered — don't rely on order
s = {3, 1, 2}
print(s)   # order is not guaranteed
```
