## 📌 What is it?
`*args` collects extra positional arguments into a tuple. `**kwargs` collects extra keyword arguments into a dict.

## 🧠 Why AI Engineers need this
Flexible API wrappers, config builders, and decorator patterns all rely on `*args` and `**kwargs` to forward arguments transparently.

````python
# ── *args ──
def average(*args):
    return sum(args) / len(args)

print(average(0.87, 0.91, 0.89))        # 3 values
print(average(0.91, 0.88, 0.95, 0.78)) # 4 values — same function!

# ── **kwargs ──
def llm_call(prompt, **kwargs):
    """Pass any extra params to the LLM API."""
    params = {"model": "gpt-4o", "temperature": 0.7, **kwargs}
    print(f"Calling with: {params}")

llm_call("Tell me a joke")
llm_call("Write code", temperature=0.0, max_tokens=2000, stream=True)
````

````python
# ── COMBINING: order matters! ──
# Order: positional, *args, keyword-only, **kwargs
def pipeline_step(name, *inputs, verbose=False, **config):
    print(f"Step: {name}")
    print(f"Inputs: {inputs}")
    print(f"Verbose: {verbose}")
    print(f"Config: {config}")

pipeline_step(
    "embedding",           # positional
    "doc1", "doc2",        # *args
    verbose=True,          # keyword-only
    dim=1536, batch=32     # **kwargs
)
````

````python
# ── UNPACKING WITH * and ** ──
def add(a, b, c):
    return a + b + c

args = (1, 2, 3)
print(add(*args))    # unpack tuple → add(1, 2, 3)

params = {"a": 10, "b": 20, "c": 30}
print(add(**params)) # unpack dict → add(a=10, b=20, c=30)

# Merging lists and dicts
defaults = {"model": "gpt-4o", "temperature": 0.7}
overrides = {"temperature": 0.0, "max_tokens": 500}
merged = {**defaults, **overrides}   # ** unpacking to merge
print(merged)

list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = [*list1, *list2]    # * unpacking to concatenate
print(combined)
````

````python
# ── FORWARDING ARGS (decorator pattern) ──
from functools import wraps
import time

def log_args(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args={args}, kwargs={kwargs}")
        return func(*args, **kwargs)  # forward everything transparently
    return wrapper

@log_args
def create_embedding(text, model="text-embedding-3-small", dim=1536):
    return [0.1] * dim

create_embedding("Hello world", model="text-embedding-ada-002")
````

## ⚠️ Common Mistakes
```python
# ❌ Wrong argument order
def bad(*args, positional):  # positional MUST come before *args
    pass

# ✅ Correct order
def good(positional, *args, keyword_only, **kwargs):
    pass

# ❌ Modifying kwargs dict inside function
def bad(**kwargs):
    kwargs['new_key'] = 'value'   # modifies local copy — ok actually
    # but confusing! Use explicit params for clarity
```
