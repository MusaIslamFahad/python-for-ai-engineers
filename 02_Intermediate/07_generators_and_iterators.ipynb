## 📌 What is it?
Generators produce values lazily using `yield` — one at a time, on demand — without storing everything in memory. Iterators are objects with `__iter__` and `__next__`.

## 🧠 Why AI Engineers need this
Streaming LLM responses, processing large datasets token-by-token, infinite data augmentation pipelines, and batch generation all use generators.

````python
# ── GENERATOR FUNCTION ──
def batch_generator(data, batch_size):
    """Yield mini-batches from a dataset without loading all into memory."""
    for i in range(0, len(data), batch_size):
        yield data[i : i + batch_size]

dataset = list(range(100))   # simulate 100 samples

for batch in batch_generator(dataset, batch_size=16):
    print(f"Processing batch of {len(batch)} items")
    # train_step(batch)
````

````python
# ── STREAMING LLM RESPONSES ──
def stream_response(text, chunk_size=5):
    """Simulate LLM token streaming with a generator."""
    words = text.split()
    for i in range(0, len(words), chunk_size):
        chunk = " ".join(words[i:i+chunk_size])
        yield chunk + " "

response = "The quick brown fox jumps over the lazy dog and runs away."
print("Streaming: ", end="", flush=True)
for token in stream_response(response, chunk_size=3):
    print(token, end="", flush=True)
print()
````

````python
# ── GENERATOR EXPRESSIONS ──
import sys

# Generator expression — lazy, memory efficient
scores = (x ** 2 for x in range(1_000_000))
print(f"Generator object size: {sys.getsizeof(scores)} bytes")
print(f"First 5: {[next(scores) for _ in range(5)]}")

# vs list — eager, loads all into memory
# scores_list = [x**2 for x in range(1_000_000)]  # ~8MB!
````

````python
# ── CUSTOM ITERATOR ──
class EpochIterator:
    """Iterate over epochs, shuffling data each time."""
    def __init__(self, data, epochs):
        import random
        self.data = data
        self.epochs = epochs
        self.current = 0
        self._random = random
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current >= self.epochs:
            raise StopIteration
        self.current += 1
        shuffled = self.data[:]
        self._random.shuffle(shuffled)
        return self.current, shuffled

data = [1, 2, 3, 4, 5]
for epoch, shuffled in EpochIterator(data, epochs=3):
    print(f"Epoch {epoch}: {shuffled}")
````

## ⚠️ Common Mistakes
```python
# ❌ Exhausted generators — can't reuse!
gen = (x for x in range(5))
print(list(gen))  # [0, 1, 2, 3, 4]
print(list(gen))  # [] — exhausted!

# ✅ Create a new generator each time, or convert to list first

# ❌ Sending to generator before first yield
def gen():
    x = yield
    yield x * 2
g = gen()
g.send(10)   # TypeError! must call next() first
g.send(None) # ✅ or next(g) to prime it
```
