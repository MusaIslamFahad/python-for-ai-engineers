## 📌 What is it?
Inheritance lets a class reuse and extend another class. Polymorphism lets different classes share the same interface.

## 🧠 Why AI Engineers need this
ML frameworks (PyTorch `nn.Module`, HuggingFace `PreTrainedModel`) use inheritance extensively. Writing custom models, datasets, and trainers requires understanding this pattern.

````python
# ── BASE CLASS ──
class BaseModel:
    """Abstract base for all AI models."""
    
    def __init__(self, name: str):
        self.name = name
        self._is_loaded = False
    
    def load(self):
        """Load model weights."""
        self._is_loaded = True
        print(f"{self.name}: model loaded")
    
    def predict(self, input_data):
        """Must be implemented by subclasses."""
        raise NotImplementedError("Subclasses must implement predict()")
    
    def __repr__(self):
        status = "loaded" if self._is_loaded else "not loaded"
        return f"{self.__class__.__name__}('{self.name}', {status})"


# ── SUBCLASSES ──
class TextClassifier(BaseModel):
    def __init__(self, name, num_classes):
        super().__init__(name)          # call parent __init__
        self.num_classes = num_classes
    
    def predict(self, text):            # override parent method
        return {"label": "positive", "confidence": 0.92}


class TextEmbedder(BaseModel):
    def __init__(self, name, dim):
        super().__init__(name)
        self.dim = dim
    
    def predict(self, text):
        return [0.1] * self.dim         # mock embedding


clf = TextClassifier("bert-base", num_classes=2)
emb = TextEmbedder("text-embedding-3", dim=1536)

clf.load()
print(clf.predict("Great product!"))
print(emb)
````

````python
# ── POLYMORPHISM ──
def run_inference(models, inputs):
    """Works on any model that implements predict()."""
    for model, inp in zip(models, inputs):
        result = model.predict(inp)
        print(f"{model.name}: {result}")

clf = TextClassifier("bert-base", 2)
emb = TextEmbedder("text-emb-3", 1536)

# Same function, different behavior — polymorphism!
run_inference([clf, emb], ["Hello world", "AI is cool"])
````

````python
# ── isinstance / issubclass ──
print(isinstance(clf, TextClassifier))  # True
print(isinstance(clf, BaseModel))       # True — checks inheritance chain
print(issubclass(TextClassifier, BaseModel))  # True

# Useful for type-safe dispatch
def process(model):
    if isinstance(model, TextClassifier):
        print("Running classification pipeline")
    elif isinstance(model, TextEmbedder):
        print("Running embedding pipeline")

process(clf)
process(emb)
````

## ⚠️ Common Mistakes
```python
# ❌ Forgetting super()
class Child(Parent):
    def __init__(self, x, y):
        self.x = x      # Parent.__init__ never called!

# ✅
class Child(Parent):
    def __init__(self, x, y):
        super().__init__(y)  # always call super first
        self.x = x
```
