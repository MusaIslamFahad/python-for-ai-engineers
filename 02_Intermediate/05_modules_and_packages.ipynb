## 📌 What is it?
A module is a `.py` file. A package is a directory of modules with an `__init__.py`. Python's import system lets you organize and reuse code.

## 🧠 Why AI Engineers need this
Every AI project is built from packages: NumPy, PyTorch, LangChain, OpenAI. Understanding imports helps you structure your own projects professionally.

````python
# ── IMPORT STYLES ──
import os                          # import module
import os.path                     # import submodule
from pathlib import Path           # import specific class
from collections import defaultdict, Counter  # multiple imports
import numpy as np                 # alias (convention!)
import pandas as pd                # alias (convention!)
from typing import Optional, Union, List   # type hints

# What NOT to do (pollutes namespace)
# from os import *

# Inspect what's in a module
import json
print(dir(json))             # list all attributes
print(json.__file__)         # where is this module?
````

````python
# ── __name__ == '__main__' PATTERN ──
# This is how you write scripts that can be both imported and run directly.

# In your_module.py:
def process_data(data):
    return [x * 2 for x in data]

def main():
    sample = [1, 2, 3, 4, 5]
    result = process_data(sample)
    print(f"Result: {result}")

# This block only runs when the file is executed directly
# NOT when it's imported by another module
if __name__ == "__main__":
    main()

# Demonstrate:
print(f"This module's __name__: {__name__}")
# When run directly: __main__
# When imported: the module name
````

````python
# ── STANDARD LIBRARY GEMS ──
# These come built-in — no pip needed!

from collections import defaultdict, Counter, OrderedDict, deque
from itertools import islice, chain, product, combinations
from functools import partial, lru_cache, reduce
import datetime, time, random, os, sys, re, math, copy, hashlib

# lru_cache — memoization (great for embedding cache!)
from functools import lru_cache

@lru_cache(maxsize=128)
def expensive_embedding(text: str) -> tuple:
    """Cache embeddings to avoid repeated API calls."""
    print(f"Computing embedding for: {text[:30]}...")
    return tuple([hash(text) % 100] * 10)  # mock

e1 = expensive_embedding("hello world")  # computed
e2 = expensive_embedding("hello world")  # cached!
print(expensive_embedding.cache_info())
````
