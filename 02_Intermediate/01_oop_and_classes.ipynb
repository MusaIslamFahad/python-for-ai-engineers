## 📌 What is it?
Object-Oriented Programming organizes code into classes — blueprints for objects that bundle data (attributes) and behavior (methods).

## 🧠 Why AI Engineers need this
LLM clients, model wrappers, RAG systems, and data pipelines are all built as classes. Understanding OOP is essential for reading and extending ML frameworks.

````python
# ── BASIC CLASS ──
class LLMConfig:
    """Configuration for an LLM API call."""
    
    # Class attribute (shared by all instances)
    default_model = "gpt-4o"
    
    def __init__(self, model=None, temperature=0.7, max_tokens=1000):
        """Instance initializer — called on object creation."""
        # Instance attributes (unique to each object)
        self.model = model or LLMConfig.default_model
        self.temperature = temperature
        self.max_tokens = max_tokens
    
    def __repr__(self):
        """String representation for debugging."""
        return f"LLMConfig(model={self.model!r}, temp={self.temperature})"
    
    def is_deterministic(self):
        """Check if config produces deterministic output."""
        return self.temperature == 0.0
    
    def to_dict(self):
        """Serialize to dict for API request body."""
        return {
            "model": self.model,
            "temperature": self.temperature,
            "max_tokens": self.max_tokens
        }

cfg = LLMConfig(temperature=0.0)
print(cfg)                     # LLMConfig(model='gpt-4o', temp=0.0)
print(cfg.is_deterministic())  # True
print(cfg.to_dict())
````

````python
# ── PROPERTIES (controlled attribute access) ──
class EmbeddingModel:
    def __init__(self, name, dim):
        self._name = name
        self._dim = dim
        self._loaded = False
    
    @property
    def name(self):
        return self._name
    
    @property
    def dim(self):
        return self._dim
    
    @dim.setter
    def dim(self, value):
        if value not in (128, 256, 512, 1024, 1536):
            raise ValueError(f"Invalid embedding dim: {value}")
        self._dim = value
    
    def __str__(self):
        return f"{self._name} ({self._dim}d)"

model = EmbeddingModel("text-embedding-3-small", 1536)
print(model)        # text-embedding-3-small (1536d)
model.dim = 512     # OK
print(model.dim)    # 512
````

````python
# ── CLASS vs STATIC METHODS ──
class TokenCounter:
    
    _instances = 0  # class-level counter
    
    def __init__(self):
        TokenCounter._instances += 1
        self.count = 0
    
    def add(self, n):
        """Instance method — operates on self."""
        self.count += n
    
    @classmethod
    def instance_count(cls):
        """Class method — operates on the class itself."""
        return cls._instances
    
    @staticmethod
    def estimate_cost(tokens, price_per_1k=0.002):
        """Static method — utility, no access to self or cls."""
        return tokens / 1000 * price_per_1k

counter = TokenCounter()
counter.add(500)
counter.add(300)
print(f"Tokens: {counter.count}")                      # 800
print(f"Instances: {TokenCounter.instance_count()}")   # 1
print(f"Cost: ${TokenCounter.estimate_cost(800):.4f}") # $0.0016
````

## ⚠️ Common Mistakes
```python
# ❌ Forgetting self
class Foo:
    def bar():      # Missing self!
        pass

# ❌ Mutating class attributes via instance
class Foo:
    data = []       # shared!
    def add(self, x):
        self.data.append(x)  # modifies class attr

# ✅ Use instance attribute
class Foo:
    def __init__(self):
        self.data = []   # each instance gets its own list
```
