## 📌 What is it?
Python can read/write text files, binary files, JSON, and CSV. The `pathlib` module provides a modern, OOP interface to the filesystem.

## 🧠 Why AI Engineers need this
Loading datasets, saving model configs, reading JSON API responses, writing CSV results, and managing model checkpoints all require file I/O.

````python
# ── TEXT FILES ──
from pathlib import Path

# Write
p = Path("/tmp/test_output.txt")
p.write_text("Hello from AI Engineer!\nLine 2\n")

# Read all at once
content = p.read_text()
print(content)

# Read line by line (memory efficient for large files)
with open(p) as f:
    for line in f:
        print(line.strip())
````

````python
# ── JSON (the AI engineer's bread and butter) ──
import json
from pathlib import Path

config = {
    "model": "gpt-4o",
    "temperature": 0.7,
    "system_prompt": "You are a helpful AI assistant.",
    "tools": ["web_search", "code_interpreter"]
}

# Write JSON
p = Path("/tmp/config.json")
with open(p, 'w') as f:
    json.dump(config, f, indent=2)

# Read JSON
with open(p) as f:
    loaded = json.load(f)
print(loaded["model"])   # gpt-4o

# JSON to/from string
json_str = json.dumps(config, indent=2)
config2 = json.loads(json_str)
print(type(config2))   # dict
````

````python
# ── CSV ──
import csv
from pathlib import Path

# Write CSV
results = [
    {"model": "GPT-4o", "accuracy": 0.91, "latency_ms": 850},
    {"model": "Claude-3.5", "accuracy": 0.89, "latency_ms": 620},
    {"model": "Gemini-Pro", "accuracy": 0.87, "latency_ms": 720},
]

p = Path("/tmp/results.csv")
with open(p, 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=["model", "accuracy", "latency_ms"])
    writer.writeheader()
    writer.writerows(results)

# Read CSV
with open(p) as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['model']}: acc={row['accuracy']}")
````

````python
# ── PATHLIB (modern, cross-platform) ──
from pathlib import Path

# Build paths safely
root = Path("/tmp/my_project")
data_dir = root / "data"
models_dir = root / "models"

# Create directories
data_dir.mkdir(parents=True, exist_ok=True)
models_dir.mkdir(parents=True, exist_ok=True)

# Write a sample file
(data_dir / "sample.txt").write_text("sample data")

# List files
print(list(root.rglob("*.txt")))  # all txt files recursively

# File info
p = data_dir / "sample.txt"
print(p.name)      # sample.txt
print(p.stem)      # sample
print(p.suffix)    # .txt
print(p.exists())  # True
print(p.stat().st_size)  # file size in bytes
````

## ⚠️ Common Mistakes
```python
# ❌ String path concatenation
path = base_dir + "/" + filename   # breaks on Windows!

# ✅ Use pathlib
path = Path(base_dir) / filename

# ❌ Forgetting to close files
f = open("file.txt")
data = f.read()
# if error occurs here, file is never closed!

# ✅ Always use with
with open("file.txt") as f:
    data = f.read()
```
