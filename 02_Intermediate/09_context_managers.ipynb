## 📌 What is it?
Context managers handle setup and teardown automatically using the `with` statement. They implement `__enter__` and `__exit__` methods.

## 🧠 Why AI Engineers need this
Managing GPU memory, database connections, temporary files, API sessions, and model inference contexts all benefit from context managers.

````python
# ── THE with STATEMENT ──
# You already use this with files:
with open("/tmp/test.txt", "w") as f:
    f.write("hello")
# File is automatically closed here, even if an exception occurs!

# ── CUSTOM CONTEXT MANAGER (class-based) ──
import time

class Timer:
    """Context manager to time a code block."""
    def __init__(self, label=""):
        self.label = label
    
    def __enter__(self):
        self.start = time.perf_counter()
        return self   # assigned to 'as' variable
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        self.elapsed = time.perf_counter() - self.start
        print(f"⏱️ {self.label}: {self.elapsed:.4f}s")
        return False  # False = don't suppress exceptions

with Timer("model inference") as t:
    time.sleep(0.1)
    result = "generated text"

print(f"Elapsed stored: {t.elapsed:.4f}s")
````

````python
# ── contextlib.contextmanager (simpler!) ──
from contextlib import contextmanager
import os

@contextmanager
def temp_directory():
    """Create a temp dir and clean it up on exit."""
    import tempfile, shutil
    tmpdir = tempfile.mkdtemp()
    print(f"Created temp dir: {tmpdir}")
    try:
        yield tmpdir   # the 'with' block runs here
    finally:
        shutil.rmtree(tmpdir)   # always clean up
        print(f"Cleaned up: {tmpdir}")

with temp_directory() as d:
    # Create some temp files
    path = os.path.join(d, "data.json")
    with open(path, "w") as f:
        f.write('{"key": "value"}')
    print(f"Working in: {d}")
    print(f"File exists: {os.path.exists(path)}")

print(f"Directory deleted: {not os.path.exists(d)}")
````

````python
# ── PRACTICAL: GPU Memory Context Manager ──
from contextlib import contextmanager

@contextmanager
def inference_mode():
    """
    Simulate torch.no_grad() context manager pattern.
    In real code: use torch.no_grad() or torch.inference_mode()
    """
    print("Disabling gradient computation...")
    # torch.set_grad_enabled(False)
    try:
        yield
    finally:
        print("Re-enabling gradient computation...")
        # torch.set_grad_enabled(True)

with inference_mode():
    print("Running inference on 1000 samples...")
    # model(inputs)  # no gradients tracked — saves memory!
````

## ⚠️ Common Mistakes
```python
# ❌ Not using context manager for resources
f = open('huge_file.txt')
process(f)   # if process() raises, file never closes!

# ✅ Always use with
with open('huge_file.txt') as f:
    process(f)

# ❌ Returning True from __exit__ suppresses exceptions
def __exit__(self, exc_type, exc_val, exc_tb):
    return True   # silently swallows all exceptions!

# ✅ Return False (or nothing) unless intentionally suppressing
```
