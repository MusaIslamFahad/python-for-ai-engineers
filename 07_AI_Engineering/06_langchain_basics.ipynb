## 📌 What is it?
LangChain is a framework for building LLM applications — it provides abstractions for chains, prompts, memory, output parsers, and tool use.

## 🧠 Why AI Engineers need this
LangChain speeds up prototyping of LLM apps. Its `LCEL` (LangChain Expression Language) lets you compose chains declaratively. Many production RAG stacks use LangChain.

```python
# ── LANGCHAIN CORE COMPONENTS ──
# pip install langchain langchain-openai langchain-community

langchain_basics = '''
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
import os

# ── 1. LLM ──
llm = ChatOpenAI(model="gpt-4o", temperature=0.7)

# Direct invocation
response = llm.invoke([HumanMessage(content="What is RAG?")])
print(response.content)

# ── 2. PROMPT TEMPLATE ──
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {domain}."),
    ("user",   "Explain {concept} in simple terms.")
])

# ── 3. LCEL CHAIN (Prompt | LLM | Parser) ──
chain = prompt | llm | StrOutputParser()
result = chain.invoke({"domain": "AI engineering", "concept": "RAG"})
print(result)

# ── 4. STREAMING ──
for chunk in chain.stream({"domain": "Python", "concept": "generators"}):
    print(chunk, end="", flush=True)
'''

print("LangChain basics pattern:")
print(langchain_basics)
```

```python
# ── LANGCHAIN OUTPUT PARSERS ──
output_parser_code = '''
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from typing import List

# Define your expected output schema
class ModelEvaluation(BaseModel):
    model_name: str      = Field(description="Name of the model")
    score: float         = Field(description="Score 0.0-1.0")
    strengths: List[str] = Field(description="Key strengths")
    weaknesses: List[str]= Field(description="Key weaknesses")
    verdict: str         = Field(description="Overall verdict")

parser = JsonOutputParser(pydantic_object=ModelEvaluation)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an AI model evaluator. {format_instructions}"),
    ("user", "Evaluate {model_name} for {use_case}")
])

llm = ChatOpenAI(model="gpt-4o")
chain = prompt | llm | parser

result = chain.invoke({
    "model_name": "GPT-4o",
    "use_case": "code generation",
    "format_instructions": parser.get_format_instructions()
})

print(f"Model:     {result['model_name']}")
print(f"Score:     {result['score']}")
print(f"Verdict:   {result['verdict']}")
'''

print("LangChain output parsing pattern:")
print(output_parser_code)
```

```python
# ── CONVERSATION MEMORY ──
memory_code = '''
from langchain.memory import ConversationBufferWindowMemory
from langchain.chains import ConversationChain
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o", temperature=0.7)

# Window memory — keeps last K turns
memory = ConversationBufferWindowMemory(k=5, return_messages=True)

chain = ConversationChain(llm=llm, memory=memory, verbose=False)

# Multi-turn conversation
responses = [
    chain.predict(input="My name is Alex and I am learning RAG."),
    chain.predict(input="What topic did I say I was learning?"),  # remembers!
    chain.predict(input="Explain chunking strategies for RAG."),
]

for r in responses:
    print(r[:100])
    print("---")
'''

print("LangChain memory pattern:")
print(memory_code)
```
