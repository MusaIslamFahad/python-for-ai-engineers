## 📌 What is it?
Keras is a high-level API (now included in TensorFlow 2.x) that makes building and training neural networks fast and intuitive.

## 🧠 Why AI Engineers need this
Keras is widely used in production ML, Google's ecosystem, and many pre-trained models on TensorFlow Hub. Understanding both PyTorch and TF/Keras is a strong advantage.

```python
# ── KERAS SEQUENTIAL API ──
# pip install tensorflow

# Note: In this notebook we show the code patterns.
# Run in an environment where tensorflow is installed.

keras_code = '''
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(f"TensorFlow: {tf.__version__}")
print(f"Keras: {keras.__version__}")

# Sequential model — stack layers
model = keras.Sequential([
    layers.Dense(256, activation="relu", input_shape=(128,)),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(2, activation="softmax")
])

model.summary()

# Compile — define optimizer, loss, metrics
model.compile(
    optimizer=keras.optimizers.AdamW(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Fit — trains automatically!
import numpy as np
X_train = np.random.randn(500, 128)
y_train = (X_train[:, 0] > 0).astype(int)

history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    callbacks=[
        keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=2)
    ],
    verbose=1
)
'''
print("Keras Sequential API code pattern shown above.")
print("Copy and run in a TensorFlow environment.")
```

```python
# ── KERAS FUNCTIONAL API — for complex architectures ──
functional_code = '''
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Functional API — allows multiple inputs/outputs, branching

inputs = keras.Input(shape=(128,), name="embedding_input")

# Shared trunk
x = layers.Dense(256, activation="relu")(inputs)
x = layers.LayerNormalization()(x)
x = layers.Dropout(0.3)(x)

# Task-specific heads (multi-task learning)
classification = layers.Dense(3, activation="softmax", name="sentiment")(x)
regression     = layers.Dense(1, name="score")(x)

model = keras.Model(inputs=inputs, outputs=[classification, regression])
model.summary()

model.compile(
    optimizer="adam",
    loss={
        "sentiment": "sparse_categorical_crossentropy",
        "score":     "mse"
    },
    loss_weights={"sentiment": 1.0, "score": 0.5},
    metrics={"sentiment": "accuracy"}
)
'''
print("Keras Functional API pattern for multi-output models shown above.")
```
