```python
import torch
import numpy as np

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ── TENSORS — the core data structure ──
# From Python
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([[1,2,3],[4,5,6]], dtype=torch.float32)

# Factory functions
zeros = torch.zeros(3, 4)
ones  = torch.ones(2, 3)
rand  = torch.randn(4, 4)        # normal distribution
eye   = torch.eye(3)             # identity matrix
arange = torch.arange(0, 10, 2)  # [0, 2, 4, 6, 8]

print(f"Shape: {b.shape}")   # torch.Size([2, 3])
print(f"Dtype: {b.dtype}")   # torch.float32
print(f"Device: {b.device}") # cpu
```

```python
import torch
import numpy as np

# ── TENSOR OPERATIONS ──
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

print(a + b)             # element-wise add
print(a @ b)             # matrix multiply
print(a.T)               # transpose
print(a.sum())           # sum all elements
print(a.mean(dim=0))     # mean along rows
print(a.max())

# ── NumPy interop ──
np_arr = np.array([1.0, 2.0, 3.0])
tensor = torch.from_numpy(np_arr)    # shares memory!
back   = tensor.numpy()              # back to numpy

print(f"numpy → tensor: {tensor}")
print(f"tensor → numpy: {back}")

# ── GPU (if available) ──
x = torch.randn(3, 3)
x_gpu = x.to(device)   # move to GPU
x_cpu = x_gpu.cpu()    # move back to CPU
print(f"On device: {x_gpu.device}")
```

```python
import torch

# ── AUTOGRAD — automatic differentiation ──
# This is how neural networks learn!
x = torch.tensor([2.0], requires_grad=True)

# Forward pass: y = x^3 + 2x^2 + 5
y = x**3 + 2*x**2 + 5

# Backward pass: compute gradients
y.backward()

# Gradient: dy/dx = 3x^2 + 4x = 3(4) + 4(2) = 12 + 8 = 20
print(f"y = {y.item():.2f}")
print(f"dy/dx = {x.grad.item():.2f}")   # 20.0

# ── torch.no_grad() — inference without tracking ──
with torch.no_grad():
    val = x**3 + 2*x**2 + 5
    print(f"No grad: {val.item():.2f}")
    print(f"Requires grad: {val.requires_grad}")  # False
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# ── CUSTOM DATASET ──
class TextDataset(Dataset):
    def __init__(self, texts, labels, max_len=10):
        self.texts  = texts
        self.labels = labels
        self.max_len = max_len
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        # Tokenize (simplified — real code uses a tokenizer)
        tokens = [hash(w) % 1000 for w in self.texts[idx].split()]
        # Pad / truncate
        tokens = tokens[:self.max_len] + [0] * max(0, self.max_len - len(tokens))
        return {
            "input_ids": torch.tensor(tokens, dtype=torch.long),
            "label":     torch.tensor(self.labels[idx], dtype=torch.long)
        }

texts  = ["machine learning is great", "deep learning needs data",
          "transformers changed nlp", "python is the language of ai"]
labels = [1, 1, 1, 0]

dataset = TextDataset(texts, labels)
loader  = DataLoader(dataset, batch_size=2, shuffle=True)

for batch in loader:
    print(f"input_ids shape: {batch['input_ids'].shape}")
    print(f"labels:          {batch['label']}")
    break
```
