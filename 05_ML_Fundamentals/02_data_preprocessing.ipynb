## 📌 What is it?
Preprocessing transforms raw data into a format that ML models can consume — scaling, encoding, imputation, and feature engineering.

## 🧠 Why AI Engineers need this
Garbage in, garbage out. Preprocessing is often 70% of the ML work. Even LLM fine-tuning requires text cleaning, tokenization, and data collation.

```python
import pandas as pd
import numpy as np

# Sample dataset with mixed types
df = pd.DataFrame({
    "age":       [25, 32, None, 45, 28, None, 35],
    "income":    [50000, 75000, 60000, None, 48000, 90000, None],
    "city":      ["NYC", "LA", "NYC", "Chicago", None, "LA", "Chicago"],
    "education": ["Bachelor", "Master", "PhD", "Bachelor", "Master", "PhD", "Bachelor"],
    "label":     [0, 1, 1, 0, 0, 1, 1]
})

print(df)
print(f"
Missing values:
{df.isnull().sum()}")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "age":       [25, 32, None, 45, 28, None, 35],
    "income":    [50000, 75000, 60000, None, 48000, 90000, None],
    "city":      ["NYC", "LA", "NYC", "Chicago", None, "LA", "Chicago"],
    "education": ["Bachelor", "Master", "PhD", "Bachelor", "Master", "PhD", "Bachelor"],
})

# ── NUMERIC: Impute then Scale ──
num_cols = ["age", "income"]
imputer = SimpleImputer(strategy="median")
df[num_cols] = imputer.fit_transform(df[num_cols])

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])
print("Scaled numeric features:")
print(df_scaled[num_cols].round(3))

# ── CATEGORICAL: Impute then Encode ──
cat_imputer = SimpleImputer(strategy="most_frequent")
df["city"] = cat_imputer.fit_transform(df[["city"]]).ravel()

ohe = OneHotEncoder(sparse_output=False, drop='first')
city_encoded = ohe.fit_transform(df[["city"]])
print(f"
City encoded shape: {city_encoded.shape}")
print(f"Categories: {ohe.categories_}")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import numpy as np

# ── SCALING COMPARISON ──
data = np.array([[1000], [2000], [3000], [100000]])  # income with outlier

standard = StandardScaler().fit_transform(data)
minmax   = MinMaxScaler().fit_transform(data)
robust   = RobustScaler().fit_transform(data)    # robust to outliers

print("Original → StandardScaler → MinMaxScaler → RobustScaler")
for i in range(len(data)):
    print(f"{data[i][0]:8.0f} → {standard[i][0]:8.3f} → {minmax[i][0]:.3f} → {robust[i][0]:.3f}")

print("StandardScaler: mean=0, std=1. Good default for most algorithms.")
print("MinMaxScaler:   range [0,1]. Good for neural networks.")
print("RobustScaler:   uses median/IQR. Best when outliers are present.")
```
