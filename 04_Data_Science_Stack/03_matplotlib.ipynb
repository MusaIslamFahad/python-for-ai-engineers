## 📌 What is it?
Matplotlib is Python's foundational plotting library. Every AI/ML visualization eventually builds on it.

## 🧠 Why AI Engineers need this
Loss curves, confusion matrices, attention heatmaps, benchmark comparisons, embedding visualizations — all built with Matplotlib.

```python
import matplotlib
matplotlib.use('Agg')   # non-interactive backend (for scripts)
import matplotlib.pyplot as plt
import numpy as np

# ── LINE PLOT — Training Loss Curve ──
epochs = np.arange(1, 21)
train_loss = 2.5 * np.exp(-0.2 * epochs) + np.random.randn(20) * 0.05
val_loss   = 2.5 * np.exp(-0.18 * epochs) + np.random.randn(20) * 0.08 + 0.1

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(epochs, train_loss, 'b-o', linewidth=2, markersize=5, label='Train Loss')
ax.plot(epochs, val_loss,   'r--s', linewidth=2, markersize=5, label='Val Loss')

ax.set_xlabel("Epoch", fontsize=12)
ax.set_ylabel("Loss", fontsize=12)
ax.set_title("Training & Validation Loss", fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_xlim(1, 20)

plt.tight_layout()
plt.savefig("/tmp/loss_curve.png", dpi=150)
plt.close()
print("✅ Saved loss_curve.png")
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── SUBPLOTS — Multiple charts in one figure ──
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1) Bar chart — model benchmark
models   = ["GPT-4o", "Claude", "Gemini", "Llama"]
accuracy = [0.912, 0.895, 0.878, 0.851]
colors   = ['#4C72B0','#DD8452','#55A868','#C44E52']

axes[0].bar(models, accuracy, color=colors, edgecolor='white', linewidth=1.2)
axes[0].set_ylim(0.80, 0.95)
axes[0].set_title("Model Accuracy", fontweight='bold')
axes[0].set_ylabel("Accuracy")
axes[0].tick_params(axis='x', rotation=15)
for i, v in enumerate(accuracy):
    axes[0].text(i, v + 0.002, f'{v:.3f}', ha='center', fontsize=9)

# 2) Scatter — accuracy vs latency
latency = [850, 620, 720, 420]
axes[1].scatter(latency, accuracy, c=colors, s=150, edgecolors='black', linewidth=1)
for i, m in enumerate(models):
    axes[1].annotate(m, (latency[i], accuracy[i]),
                     textcoords="offset points", xytext=(5,5), fontsize=8)
axes[1].set_xlabel("Latency (ms)")
axes[1].set_ylabel("Accuracy")
axes[1].set_title("Accuracy vs Latency", fontweight='bold')

# 3) Histogram — score distribution
scores = np.random.beta(8, 2, 500)
axes[2].hist(scores, bins=30, color='#4C72B0', edgecolor='white', alpha=0.85)
axes[2].set_xlabel("Score")
axes[2].set_ylabel("Count")
axes[2].set_title("Score Distribution", fontweight='bold')
axes[2].axvline(scores.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {scores.mean():.2f}')
axes[2].legend()

plt.suptitle("LLM Benchmark Dashboard", fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig("/tmp/benchmark_dashboard.png", dpi=150, bbox_inches='tight')
plt.close()
print("✅ Saved benchmark_dashboard.png")
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

# ── CONFUSION MATRIX ──
from matplotlib.colors import LinearSegmentedColormap

labels = ["Positive", "Negative", "Neutral"]
cm = np.array([[120, 15, 8],
               [12,  98, 6],
               [5,   9, 87]])

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(cm, cmap='Blues')
plt.colorbar(im, ax=ax)

ax.set_xticks(range(len(labels))); ax.set_xticklabels(labels)
ax.set_yticks(range(len(labels))); ax.set_yticklabels(labels)
ax.set_xlabel("Predicted"); ax.set_ylabel("Actual")
ax.set_title("Confusion Matrix — Sentiment Classifier", fontweight='bold')

for i in range(len(labels)):
    for j in range(len(labels)):
        color = "white" if cm[i,j] > cm.max()/2 else "black"
        ax.text(j, i, cm[i,j], ha='center', va='center', color=color, fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig("/tmp/confusion_matrix.png", dpi=150)
plt.close()
print("✅ Saved confusion_matrix.png")
```
