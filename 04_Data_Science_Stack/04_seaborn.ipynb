## 📌 What is it?
Seaborn is a statistical visualization library built on Matplotlib. It produces beautiful plots with less code.

## 🧠 Why AI Engineers need this
Correlation heatmaps, distribution comparisons, feature relationship plots, and evaluation visualizations are all easier with Seaborn.

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid", palette="muted")
print(f"Seaborn version: {sns.__version__}")

# ── DATASET ──
np.random.seed(42)
n = 200
df = pd.DataFrame({
    "accuracy":    np.random.beta(8, 2, n),
    "latency_ms":  np.random.gamma(5, 80, n),
    "cost":        np.random.exponential(0.002, n),
    "model_type":  np.random.choice(["GPT", "Claude", "Gemini"], n),
    "task":        np.random.choice(["QA", "Coding", "Summarize"], n),
})

# ── DISTRIBUTION PLOT ──
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df["accuracy"], bins=25, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title("Accuracy Distribution")

sns.boxplot(data=df, x="model_type", y="accuracy", ax=axes[1],
            palette=["#4C72B0","#DD8452","#55A868"])
axes[1].set_title("Accuracy by Model Type")

plt.tight_layout()
plt.savefig("/tmp/seaborn_distributions.png", dpi=150)
plt.close()
print("✅ Saved seaborn_distributions.png")
```

```python
import matplotlib
matplotlib.use('Agg')
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ── HEATMAP — Correlation Matrix ──
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 5),
                  columns=["BLEU","ROUGE","BERTScore","Human","GPT-Eval"])
# make correlated
df["GPT-Eval"] = df["Human"] * 0.8 + np.random.randn(100) * 0.2
df["BERTScore"] = df["Human"] * 0.6 + df["BLEU"] * 0.3 + np.random.randn(100) * 0.1

fig, ax = plt.subplots(figsize=(7, 6))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm",
            vmin=-1, vmax=1, ax=ax, linewidths=0.5)
ax.set_title("Metric Correlation Heatmap", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig("/tmp/heatmap.png", dpi=150)
plt.close()
print("✅ Saved heatmap.png")
```
