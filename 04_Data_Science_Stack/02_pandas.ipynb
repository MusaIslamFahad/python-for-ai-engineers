## 📌 What is it?
Pandas provides `DataFrame` (2D table) and `Series` (1D labeled array) for data manipulation and analysis.

## 🧠 Why AI Engineers need this
Loading training data, EDA, feature engineering, cleaning LLM evaluation results, and building benchmark leaderboards all use Pandas.

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "model":       ["GPT-4o", "Claude-3.5", "Gemini-Pro", "Llama-3", "Mistral"],
    "provider":    ["OpenAI", "Anthropic", "Google", "Meta", "Mistral AI"],
    "accuracy":    [0.912, 0.895, 0.878, 0.851, 0.834],
    "latency_ms":  [850, 620, 720, 420, 380],
    "cost_per_1k": [0.0025, 0.003, 0.0015, 0.0002, 0.0001],
    "context_k":   [128, 200, 32, 128, 32]
})

# ── SELECTING DATA ──
print(df["model"])                    # Series
print(df[["model", "accuracy"]])      # DataFrame (list of cols)
print(df.iloc[0])                     # row by integer position
print(df.loc[df["accuracy"] > 0.88])  # boolean filtering

# Chained filtering
top = df[(df["accuracy"] > 0.85) & (df["latency_ms"] < 700)]
print(f"
Fast + accurate models:
{top[['model','accuracy','latency_ms']]}") 
````

````python
import pandas as pd

df = pd.DataFrame({
    "model":       ["GPT-4o", "Claude-3.5", "Gemini-Pro", "Llama-3", "Mistral"],
    "provider":    ["OpenAI", "Anthropic", "Google", "Meta", "Mistral AI"],
    "accuracy":    [0.912, 0.895, 0.878, 0.851, 0.834],
    "latency_ms":  [850, 620, 720, 420, 380],
    "cost_per_1k": [0.0025, 0.003, 0.0015, 0.0002, 0.0001],
    "context_k":   [128, 200, 32, 128, 32]
})

# ── AGGREGATION & GROUPBY ──
print(df[["accuracy","latency_ms"]].describe())

# Group by provider (useful for multi-model benchmarks)
by_provider = df.groupby("provider").agg({
    "accuracy":    "mean",
    "latency_ms":  "min",
    "cost_per_1k": "mean"
}).round(4)
print(f"
By provider:
{by_provider}")

# Add computed columns
df["value_score"] = df["accuracy"] / (df["cost_per_1k"] * 1000)
df["grade"] = df["accuracy"].apply(lambda x: "A" if x > 0.9 else "B" if x > 0.85 else "C")
print(f"
With derived cols:
{df[['model','grade','value_score']].sort_values('value_score', ascending=False)}")
````

````python
import pandas as pd
import numpy as np

# ── HANDLING MISSING DATA ──
eval_df = pd.DataFrame({
    "model":    ["GPT-4o", "Claude", "Gemini", "Llama"],
    "task_a":   [0.92, None, 0.88, 0.85],
    "task_b":   [0.89, 0.91, None, 0.82],
    "task_c":   [None, 0.87, 0.90, None],
})

print(eval_df)
print(f"
Missing values:
{eval_df.isnull().sum()}")

# Fill or drop
filled = eval_df.fillna(eval_df.mean(numeric_only=True))
dropped = eval_df.dropna()

print(f"
Filled:
{filled}")
print(f"
Dropped rows with NaN:
{dropped}")
````

## ⚠️ Common Mistakes
```python
# ❌ Chained assignment — may not modify original
df[df['x'] > 0]['y'] = 1   # SettingWithCopyWarning!

# ✅ Use .loc
df.loc[df['x'] > 0, 'y'] = 1

# ❌ Iterating rows with iterrows() (very slow!)
for idx, row in df.iterrows():
    result = compute(row['col'])

# ✅ Vectorized apply or direct operations
df['result'] = df['col'].apply(compute)  # still slow
df['result'] = compute_vectorized(df['col'].values)  # fastest
```
